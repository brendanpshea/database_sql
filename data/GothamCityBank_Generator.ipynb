```python
%%sql
DROP TABLE IF EXISTS Customers;
DROP TABLE IF EXISTS Accounts;

-- Create Customers table
CREATE TABLE Customers (
    CustomerID INTEGER PRIMARY KEY AUTOINCREMENT,
    FirstName VARCHAR(50) NOT NULL,
    LastName VARCHAR(50) NOT NULL,
    MiddleInitial CHAR(1),
    Address VARCHAR(255),
    City VARCHAR(50),
    State VARCHAR(50),
    ZipCode VARCHAR(10),
    PhoneNumber VARCHAR(15),
    Email VARCHAR(100)
);

-- Create Accounts table
CREATE TABLE Accounts (
    AccountID INTEGER PRIMARY KEY AUTOINCREMENT,
    CustomerID INT,
    AccountType VARCHAR(20),
    Balance DECIMAL(15, 2) DEFAULT 0.00,
    CreatedDate DATE,
    FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID)
);

```

```python
import sqlite3
import random
import string
from datetime import datetime, timedelta

# Connect to the database (or create it if it doesn't exist)
conn = sqlite3.connect('gotham.db')
c = conn.cursor()

# Create tables if they don't exist
c.execute('''CREATE TABLE IF NOT EXISTS Customers (
    CustomerID INTEGER PRIMARY KEY AUTOINCREMENT,
    FirstName TEXT NOT NULL,
    LastName TEXT NOT NULL,
    MiddleInitial TEXT,
    Address TEXT,
    City TEXT,
    State TEXT,
    ZipCode TEXT,
    PhoneNumber TEXT,
    Email TEXT
)''')

c.execute('''CREATE TABLE IF NOT EXISTS Accounts (
    AccountID INTEGER PRIMARY KEY AUTOINCREMENT,
    CustomerID INTEGER,
    AccountType TEXT,
    Balance REAL DEFAULT 0.00,
    CreatedDate TEXT,
    FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID)
)''')

# Lists of names and addresses
first_names = ["Bruce", "Clark", "Diana", "Barry", "Hal", "Oliver", "Dinah", "Arthur", "Victor", "Billy",
               "Peter", "Tony", "Steve", "Natasha", "Thor", "Loki", "Wanda", "Vision", "T'Challa", "Stephen", "Matt", "Frank",
               "Jessica", "Luke", "Danny", "Elektra", "Wade", "Logan", "Scott", "Jean", "Ororo", "Rogue", "Charles", "Erik",
               "Raven", "Hank", "Warren", "Bobby", "Kitty", "Kurt", "Reed", "Susan", "Johnny", "Ben", "Hellboy", "Liz", "Abe",
               "Johann", "Roger"]
last_names = ["Wayne", "Kent", "Prince", "Allen", "Jordan", "Queen", "Lance", "Curry", "Stone", "Batson", "Parker", "Stark",
              "Rogers", "Romanoff", "Odinson", "Laufeyson", "Maximoff", "Shade", "Udaku", "Strange", "Murdock", "Castle",
              "Jones", "Cage", "Rand", "Natchios", "Wilson", "Howlett", "Summers", "Grey", "Munroe", "LeBeau", "Xavier",
              "Lehnsherr", "Darkholme", "McCoy", "Worthington", "Drake",
              "Pryde", "Wagner", "Richards", "Storm", "Storm", "Grimm", "Bruttenholm", "Sherman", "Sapien", "Krauss", "Homunculus"]

middle_initials = list(string.ascii_uppercase)
streets = ["Mountain Drive", "Kent Lane", "Prince Street", "Allen Avenue", "Jordan Boulevard", "Curry Road", "Stone Street", "Queen Avenue", "Zor-El Lane", "Kyle Street", "Lane Road", "Quinn Drive", "Isley Avenue", "Crane Road", "Luthor Boulevard", "Drake Street"]

# Dictionary of cities with corresponding states and zip codes
locations = {
    "Gotham": ("NY", ["10001", "10002", "10003", "10004"]),
    "Blüdhaven": ("NJ", ["07001", "07002", "07003", "07004"]),
    "Metropolis": ("NY", ["10101", "10102", "10103", "10104"]),
    "Star City": ("CA", ["90001", "90002", "90003", "90004"])
}
phone_prefixes = ["123", "234", "345", "456", "567", "678", "789", "890"]

def random_date(start, end):
    return start + timedelta(days=random.randint(0, int((end - start).days)))

def generate_balance():
    # Generate balance using triangular distribution skewed towards lower values
    return round(random.triangular(10, 50000, 7500), 2)

def create_customers_and_accounts(num_customers=10000, num_accounts=15000):
    # Insert customers
    for _ in range(num_customers):
        first_name = random.choice(first_names)
        last_name = random.choice(last_names)
        middle_initial = random.choice(middle_initials)
        address = f"{random.randint(1, 9999)} {random.choice(streets)}"

        # Make Gotham more common
        if random.random() < 0.5:
            city = "Gotham"
        else:
            city = random.choice(list(locations.keys()))

        state, zip_codes = locations[city]
        zip_code = random.choice(zip_codes)
        phone_number = f"{random.choice(phone_prefixes)}-{random.randint(100, 999)}-{random.randint(1000, 9999)}"

        # choose email suffix from list
        email_suffixes = ["gotham.edu", "wayne.com", "gmail.com", "yahoo.com", "outlook.com", "hotmail.com"]
        email_suffix = random.choice(email_suffixes)
        email = f"{first_name.lower()}.{last_name.lower()}@{email_suffix}"

        c.execute('''INSERT INTO Customers (FirstName, LastName, MiddleInitial, Address, City, State, ZipCode, PhoneNumber, Email)
                     VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)''',
                  (first_name, last_name, middle_initial, address, city, state, zip_code, phone_number, email))

    # Commit the changes to get CustomerIDs
    conn.commit()

    # Get all customer IDs
    c.execute("SELECT CustomerID FROM Customers")
    customer_ids = [row[0] for row in c.fetchall()]

    # Insert accounts
    account_types = ["Checking", "Savings"]
    start_date = datetime.strptime('2003-01-01', '%Y-%m-%d')
    end_date = datetime.strptime('2023-12-31', '%Y-%m-%d')

    for _ in range(num_accounts):
        customer_id = random.choice(customer_ids)
        account_type = random.choice(account_types)
        balance = generate_balance()

        # if account is savings, multiply
        if account_type == "Savings":
            balance *= 1.87

        created_date = random_date(start_date, end_date).strftime('%Y-%m-%d')

        c.execute('''INSERT INTO Accounts (CustomerID, AccountType, Balance, CreatedDate)
                     VALUES (?, ?, ?, ?)''',
                  (customer_id, account_type, balance, created_date))

    # Commit the changes
    conn.commit()

# Create customers and accounts
create_customers_and_accounts()

# Close the connection
conn.close()
```
